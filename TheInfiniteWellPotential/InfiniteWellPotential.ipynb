```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
```

```python
def psi(x,n,a=1): 
    return np.where(
        (x >= 0) & (x <= a),  # Kondisi dalam well (0 ≤ x ≤ a)
        np.sqrt(2/a) * np.sin(n * np.pi * x / a),  
        0  # Di luar well, fungsi gelombang = 0
    )
x = np.linspace(-0.5,1.5,300)
y = psi(x,1) # n_1 
y1 = psi(x,2) # n_2
y2 = psi(x,3) # n_3
```

```python
plt.plot(x,1+y,label="n=1")
plt.plot(x,4+y1,label="n=2")
plt.plot(x,9+y2,label="n=3")

plt.vlines(x=0, ymin=0, ymax=11, colors='black', linestyles='dashed', label="Potential Barrier")
plt.vlines(x=1, ymin=0, ymax=11, colors='black', linestyles='dashed')

plt.xlabel('x(m)')
plt.ylabel('V(x)')
plt.title('The Infinite Well Potential')
plt.legend(fontsize=8)
plt.grid(True, linestyle='--', linewidth=0.6)

# Simpan dalam format HD
plt.savefig("InfiniteWellPotential.png", dpi=300, bbox_inches='tight', format='png')  
plt.show()
```
